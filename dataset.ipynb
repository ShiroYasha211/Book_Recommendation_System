```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('programming_books_dataset.csv')
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print("عدد القيم المفقودة فى كل عمود :")
print(df.isnull().sum())
```

```python
duplicates = df.duplicated().sum()
print(f"عدد الصفوف المكررة فى البيانات : {duplicates}")
```

```python
category_counts = df['category'].value_counts()
print("عدد الكتب فى كل فئة برمجية:\n",category_counts)
```

```python
plt.figure(figsize=(8,5))
category_counts.head(10).plot(kind='bar')
plt.title('أكثر 10 فئات برمجية شيوعاً')
plt.xlabel('الفئة البرمجية')
plt.ylabel('عدد الكتب')
plt.show()
```

```python
avg_rating = df['rating'].mean()
print(f"متوسط تقييم الكتب البرمجية : {avg_rating:.2f}")
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['pages'], df['rating'], alpha=0.6)
plt.title('العلاقة بين عدد الصفحات وتقييم الكتاب')
plt.xlabel('عدد الصفحات')
plt.ylabel('التقييم')
plt.show()
```

```python
plt.figure(figsize=(6,4))
df['rating'].hist(bins=10)
plt.title('توزيع التقييمات')
plt.xlabel('التقييم')
plt.ylabel('عدد الكتب')
plt.show()
```
